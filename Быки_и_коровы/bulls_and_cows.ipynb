**Коровы и быки**

```python
from random import randint # импортируем функцию для создания случайных чисел
answer = [] # создаём список для хранения загаданных компьютером чисел
inp = randint(0, 9) # компьютер выбирает первое число
for numeral in range(1,5): # цикл для генерирвоания 4 чисел
    while inp in answer: # проверяем, была ли уже такая цифра
        inp = randint(0,9) # если была, генерируем новую цифру
    answer.append(inp) # записываем цифру, которой еще не было
win = False # переменная для проверки
while win == False: # пока проверка не выполнилась, выполняем цикл
    repeat = True # переменная для проверки, что пользователь не ввёл одинаковые цифры
    check = '' # создаём строку для хранения введённого пользователем числа
    while len(check) != 4 or repeat == True: # выполняем цикл пока пользователь не введёт 4-х значное число, без повторений цифр
        check = input('Введите четырёхзначное число, без повторений цифр')
        incr = 1 # счетчик для отслеживания проверяемого число
        repeat = False # обнуляем переменную проверки
        for num in check: # для каждой цифры во введённом числе
            for next_num in range(incr,4): # сравниваем число с числами, идущего после него БОЛЕЕ ПОДРОБНО СМ. В ДНЕВНИЧКЕ
                if num == check[next_num]: # если цифры повторяются
                    repeat = True # записываем в переменную, что числа повторяются
                incr += 1 # увеличиваем счетчик для отслеживания
    number_bull = 0 # создаём переменную для отсчёта числа быков
    number_cow = 0 # создаём переменную для отсчёта числа коров
    incr = 0 # создаём счётчик, для отслеживания индекса проверяемой цифры
    for numeral in answer: # для каждой цифры сгенерированной компьютером
        if numeral == int(check[incr]): # если цифры на одних и тех же позициях совпали
            number_bull += 1 # увеличиваем число быклв
        incr += 1 # увеличиваем индекс проверяемой позиции
    for all in check: # каждой цифры во введённом пользователем числе
        for numeral in answer: # для каждой цифры в сгенерированном компьютером числе
            if int(all) == numeral: # если цифры совпали
                number_cow += 1 # увеличиваем число коров
    number_cow -= number_bull # из числа коров, вычитаем число быков, чтобы убрать из счёта числа корова цифры, стоящие в нужной позиции
    print(f'Число коров:{number_cow}, число быков: {number_bull}') # выводим информацию
    if number_bull == 4: # если число быков равно 4, то пользователь отгадал все 4 цифры
        win = True
print('Поздравляю, вы отгадали число')
```
