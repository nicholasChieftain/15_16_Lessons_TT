```python
import random

#Типы состояний клеток на поле для генерации и отрисовки
CELL = 0 #Пустая клетка
WALL = 1 #Стена
VISITED = 2 #Была посещена 
GO = 3 #Можно идти на эту клетку при генерации
START_POSITION = 4 #Стартовая позиция
QUES = 5 #Вопросы для реализации квеста

number_of_row_without_border = 21 #Количество строк в лабиринте без границ
number_of_column_without_border = 7 #Количество столбцов в лабиринте без границ


#Функция проверки соседей вокруг клетки, функция возвращает список клеток вокруг клетки,
#начиная слева и далее против часовой стрелки
def get_neighbors_around(main_board, current_position):
    temp_neighbors = list(range(4))
    move_x = 2
    move_y = 2
    iterator_for_neighbors = 0
    for i in range(2):
        next_x = current_position[0] + move_x * (-1) ** i
        next_y = current_position[1] + move_y * (-1) ** i

        if (0 < next_x < number_of_column_without_border + 2) and (main_board[next_x][current_position[1]] != VISITED):
            temp_neighbors[iterator_for_neighbors] = GO
        else:
            temp_neighbors[iterator_for_neighbors] = WALL
        if (0 < next_y < number_of_row_without_border + 2) and (main_board[current_position[0]][next_y] != VISITED):
            temp_neighbors[iterator_for_neighbors + 1] = GO
        else:
            temp_neighbors[iterator_for_neighbors + 1] = WALL
        iterator_for_neighbors += 2

    return temp_neighbors


inner_list_of_board = []
main_list_of_board = []

#Генерация шаблона игрового поля
for i in range(number_of_column_without_border + 2):
    if i % 2 != 1:
        inner_list_of_board = [1 for i in range(number_of_row_without_border + 2)]
    else:
        inner_list_of_board = [int(not i % 2) for i in range(number_of_row_without_border + 2)]
    print(inner_list_of_board)
    main_list_of_board.append(inner_list_of_board)

flag_done = False
current_position = [1, 1]
start_position = current_position.copy()
list_of_movement = []
stack_of_movement = []


#Цикл генерации лабиринта
while not flag_done:
    list_of_movement.append(current_position) #Сохраняем текущее положение в список положений
    neighbors = get_neighbors_around(main_list_of_board, current_position) #Определяем клетки вокруг текущей клетки
    list_for_choice = [i for i in range(4) if neighbors[i] == GO]

    if GO not in neighbors:
        stack_of_movement = list_of_movement[::-1].copy()
        main_list_of_board[stack_of_movement[0][0]][stack_of_movement[0][1]] = QUES
        for i in range(len(stack_of_movement)):
            previuos_position = stack_of_movement[i]
            neighbors = get_neighbors_around(main_list_of_board, stack_of_movement[i])
            list_of_movement.remove(previuos_position)
            list_for_choice = [i for i in range(4) if neighbors[i] == GO]
            if GO in neighbors:
                current_position = previuos_position.copy()
                break
            if previuos_position == start_position and GO not in neighbors:
                flag_done = 1
                break
    try:
        random_direction = random.choice(list_for_choice)
        main_list_of_board[current_position[0]][current_position[1]] = VISITED
        if random_direction == 0:
            main_list_of_board[current_position[0] + 1][current_position[1]] = VISITED
            current_position = [current_position[0] + 2, current_position[1]]
        elif random_direction == 1:
            main_list_of_board[current_position[0]][current_position[1] + 1] = VISITED
            current_position = [current_position[0], current_position[1] + 2]
        elif random_direction == 2:
            main_list_of_board[current_position[0] - 1][current_position[1]] = VISITED
            current_position = [current_position[0] - 2, current_position[1]]
        elif random_direction == 3:
            main_list_of_board[current_position[0]][current_position[1] - 1] = VISITED
            current_position = [current_position[0], current_position[1] - 2]
    except IndexError:
        print('Erroredn')
        print(current_position)
        print( main_list_of_board[current_position[0]][current_position[1]])
        print(random_direction)
        flag_done = 1

        
main_list_of_board[start_position[0]][start_position[1]] = 4
player_position = start_position.copy()


def draw_board(main_board):
    for row in main_board:
        for cell in row:
            if cell == WALL or cell == CELL:
                print('#', end='')
            elif cell == VISITED:
                print(' ', end='')
            elif cell == START_POSITION:
                print('O',end='')
            elif cell == QUES:
                print('?',end='')
        print()
        
def move_player(direction):
    main_list_of_board[player_position[0]][player_position[1]] = 2
    if direction == 0 and main_list_of_board[player_position[0]][player_position[1] - 1] not in [WALL, CELL]:
        player_position[1] -= 1
    elif direction == 1 and main_list_of_board[player_position[0]][player_position[1] + 1] not in [WALL, CELL]:
        player_position[1] += 1
    elif direction == 2 and main_list_of_board[player_position[0] + 1][player_position[1]] not in [WALL, CELL]:
        player_position[0] += 1
    elif direction == 3 and main_list_of_board[player_position[0] - 1][player_position[1]] not in [WALL, CELL]:
        player_position[0] -= 1
    main_list_of_board[player_position[0]][player_position[1]] = 4
    
# def check_quest(player_pos, main_board):
#     if main_board[player_pos[0]][player_pos[1]] == 
```
