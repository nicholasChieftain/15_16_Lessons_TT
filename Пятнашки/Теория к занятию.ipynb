### Список в списке

Как вы помните из занятия 9, мы можем хранить в списке всё что угодно, даже другой список. Давайте рассмотрим работу с вложенным списками

**Индексация элементов**

<img src="img/1.png" width="800">

```python
stories = [['Евгений Онегин','Капитанская дочка'],['Ревизор','Женитьба','Игроки']]
print(stories[0]) # Выводим произведения Пушкина
print(stories[1]) # Выводим произведения Гоголя
```

В коду выше, мы обращаемся к первому списку с помощью индекса 0 и получаем список произведений Пушкина. Ко второму списку мы обращаемся с помощью индекса 1 и получаем список произведений Гоголя.
Теперь давайте попробуем вывести третье произведение из списка Гоголя. Рассмотрим два варинта: с использованием переменной и напрямую.

**Вывод третьего произведения с использованием переменной**

```python
stories = [['Евгений Онегин','Капитанская дочка'],['Ревизор','Женитьба','Игроки']]
gogol = stories[1] # Записываем в переменную второй список
print(gogol[2]) # выводим третье произведение из сохраненного списка
```

Как вы видите в коде выше, мы сначала сохранили список произведений Гоголя в переменную, после чего обратились к этой переменной как к обычному списку и вывели необходимый элемент. Теперь давайте попробуем обратиться к третьему произведению без промежуточных операций.

**Вывод третьего произведения напрямую**

```python
stories = [['Евгений Онегин','Капитанская дочка'],['Ревизор','Женитьба','Игроки']]
print(stories[1][2])
```

Здесь мы сразу же обращаемся к третьему произведению. Для этого мы сначала указываем индекс списка, к которому хотим обратиться, после чего указываем индекс элемента списка, к которому обратились. Таким образом мы получаем необходимый нам элемент 

**Удаление элемента из списка**

Так же для выполнения данного задания, может пригодиться функция pop(). Данная функция применяется к массивам, она позволяет удалить элемент с указанным индексом. Если индекс не указан, то удаляется последний элемент из списка

```python
A = [2,3,4,5]
B = [10,9,8,7]
A.pop(1) # указываем индекс 1, т.е. удаляем второй элемент, тройку
B.pop() # индекс не указан, т.е. удаляем последний элемент в списке, семерку
print(A)
print(B)
```
